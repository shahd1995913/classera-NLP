```python
import requests
from bs4 import BeautifulSoup
#read 
url = "https://mea.classera.com/ar/home-ar-2/?_ga=2.68992913.1492812598.1642336142-575049594.1630744110"
r = requests.get(url)
soup=BeautifulSoup(r.content,"lxml")
#soup
firstH3 = soup.find('h3') # Start here
uls = []
for nextSibling in firstH3.findNextSiblings():
    if nextSibling.name == 'h2':
        break
    if nextSibling.name == 'ul':
        uls.append(nextSibling)
for ul in uls:
    for li in ul.findAll('li'):
        print(li)
```

```python
# page 1 
arr = []
for el in soup.find_all("ul", {"id": "nav"}):
  for a in soup.find_all("li"):
    arr.append(a.text.strip())
```
